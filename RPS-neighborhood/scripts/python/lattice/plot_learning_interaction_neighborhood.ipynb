```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

learning_diferente = [4, 8, 24, 48, 80, 120, 168]

media_frequencias = []

for l in learning_diferente:
    diretorio = f"C:\\Unicamp\\mestrado\\simulacoes\\RPS-python\\RPS-POO\\outputs\\vizinhanca_aprendizado-interacao\\teste\\learning_diferente\\vizinho_aprendizado_Y_{l}\\"
    df = pd.read_csv(os.path.join(diretorio, "frequencias.csv"))
    media = df.groupby(["t", "estrategia"])["frequencia"].mean().reset_index()
    media["learning"] = l
    media_frequencias.append(media)

media_frequencias_df = pd.concat(media_frequencias, ignore_index=True)
print(media_frequencias_df)
```

```python
ultima_geracao = media_frequencias_df[media_frequencias_df["t"] == 200]
ultima_geracao

cores = {"O": "#FD9800", "B": "#0047B3", "Y": "#FFF237"}

plt.figure(figsize=(10, 6))
plt.plot(ultima_geracao[ultima_geracao["estrategia"] == 'O']["learning"], ultima_geracao[ultima_geracao["estrategia"] == 'O']["frequencia"], marker='o', label='Estratégia O', color=cores['O'])
plt.plot(ultima_geracao[ultima_geracao["estrategia"] == 'B']["learning"], ultima_geracao[ultima_geracao["estrategia"] == 'B']["frequencia"], marker='s', label='Estratégia B', color=cores['B'])
plt.plot(ultima_geracao[ultima_geracao["estrategia"] == 'Y']["learning"], ultima_geracao[ultima_geracao["estrategia"] == 'Y']["frequencia"], marker='^', label='Estratégia Y', color=cores['Y'])

#plt.title("Frequências das estratégias ao longo do tempo")
plt.xlabel("Tamanho da vizinhança de aprendizado do Y")
plt.ylabel("Frequência em t = 200")
plt.grid(True)
plt.ylim(0, 1)
plt.savefig("C:\\Unicamp\\mestrado\\simulacoes\\RPS-python\\RPS-POO\\outputs\\vizinhanca_aprendizado-interacao\\teste\\learning_diferente\\frequencia_final_media.png", dpi=300)
plt.legend()
plt.show()
```
