```python
import sympy as smp
```

```python
t = smp.symbols('t', real=True) # determina a variável tempo (t)
fO = smp.Function('fO')(t) # determina a frequencia de O em função do tempo
fY = smp.Function('fY')(t) # determina a frequencia de Y em função do tempo
fB = 1 - fO - fY # determina a frequencia de B em função de Y e O 

b = smp.symbols('b', real =True) # determina a variável b (benefício de ganhar)
c = smp.symbols('c', real =True) # determina a variável c (custo de perder)

# monta a matriz de payoff em função de b e c
payoff_matrix = smp.Matrix([[1, b-c, b], # O
                            [b, 1, b-c], # Y
                            [b-c, b, 1]]) # B

#determina a função de fitness de cada estratégia (frequência do oponente * payoff da estratégia contra o oponente)
fitness_O = fO * payoff_matrix[0, 0] + fY * payoff_matrix[0, 1] + fB * payoff_matrix[0, 2]
fitness_Y = fO * payoff_matrix[1, 0] + fY * payoff_matrix[1, 1] + fB * payoff_matrix[1, 2]
fitness_B = fO * payoff_matrix[2, 0] + fY * payoff_matrix[2, 1] + fB * payoff_matrix[2, 2]

# determina o fitness médio da população (frequência da estratégia * fitness da estratégia)
media_fitness = fO * fitness_O + fY * fitness_Y + fB * fitness_B

# determina como se dá a variação da frequência de cada estratégia ao longo do tempo (derivada da frequência em função do tempo)
dfOdt = fO * (fitness_O - media_fitness)
dfYdt = fY * (fitness_Y - media_fitness)
dfBdt = fB * (fitness_B - media_fitness)
```

```python
# determina as frequências de equilíbrio (onde a derivada é igual a zero)
smp.solve([smp.Eq(dfOdt, 0), smp.Eq(dfYdt, 0), smp.Eq(dfBdt, 0), smp.Eq(fO + fY + fB, 1)], (fO, fY, fB))
```

```python
F = smp.Matrix([dfOdt, dfYdt])
variables = [fO, fY]
J = F.jacobian(variables)
#print(J)

J_eq = J.subs({fO: 1/3, fY: 1/3})
J_eq
autovalores = J_eq.eigenvals()
print(autovalores)
```

```python
# Pegue os autovalores como lista (caso estejam em dict)
autoval_list = list(autovalores.keys())

# Pegue a parte real dos autovalores
real_parts = [smp.re(val) for val in autoval_list]

# Imponha a condição de estabilidade (parte real < 0)
condicoes_estabilidade = [smp.Lt(r, 0) for r in real_parts]

# Resolva simbolicamente para b e c
solucao = smp.solve(condicoes_estabilidade, (b, c), dict=True)
print(solucao)
```

```python
# caso com c_O e c_B

t = smp.symbols('t', real=True) # determina a variável tempo (t)
fO = smp.Function('fO')(t) # determina a frequencia de O em função do tempo
fY = smp.Function('fY')(t) # determina a frequencia de Y em função do tempo
fB = 1 - fO - fY # determina a frequencia de B em função de Y e O 

b = smp.symbols('b', real =True, positive=True) # determina a variável b (benefício de ganhar)
c = smp.symbols('c', real =True, positive=True) # determina a variável c (custo de perder)
c_O = smp.symbols('c_O', real =True, positive=True) # determina a variável c_O (custo de jogar com O)
c_B = smp.symbols('c_B', real =True, positive=True) # determina a variável c_B (custo de jogar com B)

# monta a matriz de payoff em função de b e c
payoff_matrix = smp.Matrix([[1, b-c, b], # O
                            [b-c_O, 1, b-c_B], # Y
                            [b-c, b, 1]]) # B

#determina a função de fitness de cada estratégia (frequência do oponente * payoff da estratégia contra o oponente)
fitness_O = fO * payoff_matrix[0, 0] + fY * payoff_matrix[0, 1] + fB * payoff_matrix[0, 2]
fitness_Y = fO * payoff_matrix[1, 0] + fY * payoff_matrix[1, 1] + fB * payoff_matrix[1, 2]
fitness_B = fO * payoff_matrix[2, 0] + fY * payoff_matrix[2, 1] + fB * payoff_matrix[2, 2]

# determina o fitness médio da população (frequência da estratégia * fitness da estratégia)
media_fitness = fO * fitness_O + fY * fitness_Y + fB * fitness_B

# determina como se dá a variação da frequência de cada estratégia ao longo do tempo (derivada da frequência em função do tempo)
dfOdt = fO * (fitness_O - media_fitness)
dfYdt = fY * (fitness_Y - media_fitness)
dfBdt = fB * (fitness_B - media_fitness)

smp.solve([smp.Eq(dfOdt, 0), smp.Eq(dfYdt, 0), smp.Eq(dfBdt, 0), smp.Eq(fO + fY + fB, 1)], (fO, fY, fB))
```
